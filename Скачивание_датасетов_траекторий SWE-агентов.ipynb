```python
# scripts/download_datasets.py
from datasets import load_dataset
import json
import os

DATA_DIR = "/var/essdata/DN_1/storage/home/ttn/data/swe_trajectories"
os.makedirs(DATA_DIR, exist_ok=True)

datasets = {
    "nemotron": "nvidia/Nemotron-SWE-v1",
    "openhands": "SWE-Gym/OpenHands-Sampled-Trajectories",
    "nebius": "nebius/SWE-rebench-openhands-trajectories",
    "smith": "SWE-bench/SWE-smith-trajectories"
}

for name, repo in datasets.items():
    print("Loading", repo)
    ds = load_dataset(repo)
    # сохраните raw-версии и небольшой preview
    ds.save_to_disk(os.path.join(DATA_DIR, name))
    # экспорт первых 100 элементов в jsonl (для проверки)
    with open(os.path.join(DATA_DIR, f"{name}_preview.jsonl"), "w", encoding="utf-8") as fo:
        for i, row in enumerate(ds["train"] if "train" in ds else ds[list(ds.keys())[0]]):
            if i>=100: break
            fo.write(json.dumps(row, ensure_ascii=False) + "\n")
    print("Saved preview for", name)
```
