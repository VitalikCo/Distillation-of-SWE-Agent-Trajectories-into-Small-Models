```python
role_list = []
tool_calls_list = []
id_list = []
for i in range(len(ds["r2e_gym"]["messages"][0])):
    role = ds["r2e_gym"]["messages"][0][i]['role']
    tool_calls = ds["r2e_gym"]["messages"][0][i]['tool_calls']
    id_ = ds["r2e_gym"]["messages"][0][i]['id']
    if role not in role_list:
        role_list.append(role)
    if tool_calls not in tool_calls_list:
        tool_calls_list.append(tool_calls)
    if id_ not in id_list:
        id_list.append(id_)

print('role_list', role_list)
print('tool_calls_list', tool_calls_list)
print('id_list', id_list)
```

```python
# подробный вывод одной из траекторий (столбец messages, строка 3465) в читабельном виде

import json
BOLD = '\033[1m'
END = '\033[0m'

messages = ds["r2e_gym"]["messages"][3465]
for i in range(len(messages)):
    print('\n', '=' * 110)
    print(BOLD + 'ROLE:' + END, messages[i]['role'])
    print(BOLD + 'CONTENT:\n' + END, messages[i]['content'])
    if messages[i]['role'] == 'assistant':
        print('-' * 100)
        print(BOLD + 'TOOL_CALLS -> NAME:' + END, messages[i]['tool_calls'][0]['function']['name'])
        print(BOLD + 'TOOL_CALLS -> ARGUMENTS:' + END)
        arguments = json.loads(messages[i]['tool_calls'][0]['function']['arguments']) # json.loads('{...}') -> {...}
        for key in arguments.keys():
            print(f'{key.upper()}:', arguments[key])
```
